# Data Cleaning
---

## Load Dataset
---

```python
import os
import pandas as pd

BASE_DIR = os.path.dirname(os.getcwd())

RAW_DIR = os.path.join(BASE_DIR, "data", "raw")

customers = pd.read_csv(
    os.path.join(RAW_DIR, "customers.csv"),
    parse_dates=["signup_date", "last_login_date"]
)

subscriptions = pd.read_csv(
    os.path.join(RAW_DIR, "subscriptions.csv"),
    parse_dates=["start_date", "end_date"]
)

payments = pd.read_csv(
    os.path.join(RAW_DIR, "payments.csv"),
    parse_dates=["payment_date"]
)

usage = pd.read_csv(
    os.path.join(RAW_DIR, "product_usage.csv"),
    parse_dates=["usage_date"]
)

tickets = pd.read_csv(
    os.path.join(RAW_DIR, "support_tickets.csv"),
    parse_dates=["created_date"]
)

print("Semua data mentah berhasil dimuat tanpa error!")
```

## Referential Integrity Check
---

```python
# Payment
payments[
    ~payments["customer_id"]
    .isin(customers["customer_id"])
]
```

```python
# usage
usage[
    ~usage["customer_id"]
    .isin(customers["customer_id"])
]
```

```python
# ticket
tickets[
    ~tickets["customer_id"]
    .isin(customers["customer_id"])
]
```

```python
# subscription
subscriptions[
    ~subscriptions["customer_id"]
    .isin(customers["customer_id"])
]
```

## Business Rules Validation
---

```python
# Rule 1
subscriptions[
    subscriptions["start_date"]
    > subscriptions["end_date"]
]
```

```python
# Rule 2
payments[
    payments["amount"] <= 0
]
```

```python
# Rule 3
usage[
    usage["session_duration"] <= 0
]
```

```python
# Rule 4
customers[
    customers["age"] < 18
]
```

```python
# Rule 5
subscriptions[
    ~subscriptions["status"]
    .isin(["Active", "Churned"])
]
```

## Save Processed Data
---

```python
import os

BASE_DIR = os.path.dirname(os.getcwd())

PROCESSED_DIR = os.path.join(BASE_DIR, "data", "processed")

os.makedirs(PROCESSED_DIR, exist_ok=True)

customers.to_csv(
    os.path.join(PROCESSED_DIR, "customers_clean.csv"),
    index=False
)

subscriptions.to_csv(
    os.path.join(PROCESSED_DIR, "subscriptions_clean.csv"),
    index=False
)

payments.to_csv(
    os.path.join(PROCESSED_DIR, "payments_clean.csv"),
    index=False
)

usage.to_csv(
    os.path.join(PROCESSED_DIR, "product_usage_clean.csv"),
    index=False
)

tickets.to_csv(
    os.path.join(PROCESSED_DIR, "support_tickets_clean.csv"),
    index=False
)

print(f"Semua data bersih berhasil disimpan ke: {PROCESSED_DIR}")
```
