# Data Loading

## Load Dataset
---

```python
import os
import pandas as pd

# 1. Deteksi folder 'notebooks' tempat file ini berjalan, lalu mundur 1 tingkat ke root proyek
BASE_DIR = os.path.dirname(os.getcwd())

RAW_DIR = os.path.join(BASE_DIR, "data", "raw")

customers = pd.read_csv(
    os.path.join(RAW_DIR, "customers.csv"),
    parse_dates=["signup_date", "last_login_date"]
)

subscriptions = pd.read_csv(
    os.path.join(RAW_DIR, "subscriptions.csv"),
    parse_dates=["start_date", "end_date"]
)

payments = pd.read_csv(
    os.path.join(RAW_DIR, "payments.csv"),
    parse_dates=["payment_date"]
)

usage = pd.read_csv(
    os.path.join(RAW_DIR, "product_usage.csv"),
    parse_dates=["usage_date"]
)

tickets = pd.read_csv(
    os.path.join(RAW_DIR, "support_tickets.csv"),
    parse_dates=["created_date"]
)

print("Semua data mentah berhasil dimuat tanpa error!")
```

## Data Profiling
---

```python
def profile(df):
    return pd.DataFrame({
        "dtype": df.dtypes,
        "missing": df.isna().sum(),
        "missing_pct":
            round(df.isna().mean()*100, 2),
        "unique":
            df.nunique()
    })
```

## Shape Check
---

```python
datasets = {
    "customers": customers,
    "subscriptions": subscriptions,
    "payments": payments,
    "usage": usage,
    "tickets": tickets
}

for name, df in datasets.items():
    print(name)
    print(df.shape)
    print()
```

## Duplicated Check

```python
print("Total duplikat di Customers     :", customers.duplicated().sum())
print("Total duplikat di Subscriptions :", subscriptions.duplicated().sum())
print("Total duplikat di Payments      :", payments.duplicated().sum())
print("Total duplikat di Usage         :", usage.duplicated().sum())
print("Total duplikat di Tickets       :", tickets.duplicated().sum())
```

### Primary key check :

```python
print("Duplikat PK customer_id     :", customers["customer_id"].duplicated().sum())
print("Duplikat PK subscription_id :", subscriptions["subscription_id"].duplicated().sum())
print("Duplikat PK payment_id      :", payments["payment_id"].duplicated().sum())
print("Duplikat PK usage_id        :", usage["usage_id"].duplicated().sum())
print("Duplikat PK ticket_id       :", tickets["ticket_id"].duplicated().sum())
```

## Categorical Validation
---

```python
print("=== DISTRIBUSI KATEGORI CUSTOMERS: COUNTRY ===")
display(customers["country"].value_counts())
print("\n" + "="*50 + "\n")

print("=== DISTRIBUSI KATEGORI CUSTOMERS: ACQUISITION CHANNEL ===")
display(customers["acquisition_channel"].value_counts())
print("\n" + "="*50 + "\n")

print("=== DISTRIBUSI KATEGORI SUBSCRIPTIONS: PLAN ===")
display(subscriptions["plan"].value_counts())
print("\n" + "="*50 + "\n")

print("=== DISTRIBUSI KATEGORI PAYMENTS: PAYMENT STATUS ===")
display(payments["payment_status"].value_counts())
print("\n" + "="*50 + "\n")

print("=== DISTRIBUSI KATEGORI TICKETS: TICKET STATUS ===")
display(tickets["ticket_status"].value_counts())
```
